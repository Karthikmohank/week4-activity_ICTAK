### Male v/s Female buyers participated in sales

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(dta['Buyer Gender'])
freq_y=500
plt.yticks(np.arange(0,6000,freq_y))
plt.title('Male/Female count  participated in sales')
plt.xlabel('Gender')
plt.ylabel('Count of people')
plt.grid()
```

###  Graph of cars wrt to their sales price

```python
plt.figure(figsize=(15,10))
plt.bar(dta['Make'],dta['Sale Price'])
plt.xlabel('Brand of the car')
plt.ylabel('Sale price')
plt.title('Cars wrt their sales values')
plt.xticks(rotation=90)
plt.show()
```
